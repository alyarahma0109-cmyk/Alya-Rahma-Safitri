A. Analisalah data berikut yang berisi kumpulan data kualitas anggur yang diambil dari UCI Machine Learning Repository. Keterangan setiap baris disini berisi data tentang sampel anggur, termasuk beberapa sifat fisikokimia yang dikumpulkan dari pengujian, serta peringkat kualitas yang dievaluasi oleh para ahli anggur.
a. Lakukanlah analisis fitur tersebut sehingga menampilkan data berikut:


```python
import pandas as pd

df = pd.read_csv('winequality-red.csv')

fitur = ['alcohol','pH','residual sugar','citric acid'] # Corrected 'residual_sugar' to 'residual sugar'

for col in fitur:
    mean_val = df[col].mean()

    df[col+'_category'] = df[col].apply(
        lambda x: 'high' if x >= mean_val else 'low'
    )

    hasil = df.groupby(col+'_category')['quality'].mean()

    print(col)
    print(hasil)
```

## B. Buatlah Exploratory Data Analysis dari permasalahan tersebut.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('winequality-red.csv')

print(df.info())
print(df.describe())
print(df.isnull().sum())
```

```python
df.hist(figsize=(12,10))
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
sns.boxplot(x='quality', y='alcohol', data=df)
plt.show()
```
